```python
import altair as alt
import pandas as pd
from vega_datasets import data
```

```python
#examining the table structure
df = data.iowa_electricity()
df.head(10)
```

```python
# Seeing which companies are represented in the table
grouped_generation = df.groupby('source')['net_generation'].count()
grouped_generation
```

```python
# Create a faceted line chart
line_chart = alt.Chart(df).mark_line().encode(
    x='year:T',
    y='net_generation:Q',
    color='source:N',
    tooltip=['year:T', 'net_generation:Q']
).properties(
    width=600,
    height=300
).facet(
    column='source:N'
)

line_chart
```
