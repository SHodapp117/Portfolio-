The Pandas documentation [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#applying-multiple-functions-at-once) should be a useful reference.

```
import altair as alt
import pandas as pd


# Group the data by 'overall' and 'age', calculating the mean wage for each group
agg_data = df.groupby(['overall', 'age']).agg({'wage_eur': 'mean'}).reset_index()

# Create a stacked bar chart
stacked_bar_chart = alt.Chart(agg_data).mark_bar().encode(
    x=alt.X('overall:O'),  # Set 'overall' on the x-axis
    y=alt.Y('wage_eur:Q', stack='normalize'),  # Stack the bars and normalize the values
    color='age:N',  # Color bars by 'age'
    tooltip=['overall:O', 'wage_eur:Q', 'age:N']  # Show tooltips on hover
).properties(
    width=600,
    height=400,
    title='Stacked Bar Chart of Wages by Overall Rating and Age'  # Update the title
).configure_title(
    fontSize=16,
    anchor='start',
    offset=10
)

# Show the stacked bar chart
stacked_bar_chart
```
